```python
import os

# Root directory where your CSV files exist
data_dir = r"DataSources"    

path_data = {
    "brands":     os.path.join(data_dir, "brands.csv"),
    "categories": os.path.join(data_dir, "categories.csv"),
    "customers":  os.path.join(data_dir, "customers.csv"),
    "order_items": os.path.join(data_dir, "order_items.csv"),
    "orders":     os.path.join(data_dir, "orders.csv"),
    "products":   os.path.join(data_dir, "products.csv"),
    "staffs":     os.path.join(data_dir, "staffs.csv"),
    "stocks":     os.path.join(data_dir, "stocks.csv"),
    "stores":     os.path.join(data_dir, "stores.csv"),
}


```

## loades and df for data

```python
dfs = {}   

for name, path in path_data.items():
    dfs[name] = pd.read_csv(path)

dfs
```

## modify to lower and _ 

```python
dfs['brands']["brand_name"] = dfs['brands']["brand_name"].apply(lambda x : x.lower() .replace(" " , "_"))
dfs['categories']["category_name"] = dfs['categories']["category_name"].apply(lambda x : x.lower() .replace(" " , "_"))
dfs['products']["product_name"] = dfs['products']["product_name"].apply(lambda x : x.lower() .replace("-","_").replace(" " , "_"))

```

```python
for name, df in dfs.items():
    df.columns = [
        x.lower().replace(" ", "_") if isinstance(x, str) else x
        for x in df.columns
    ]

```

```python
cols = ["first_name", "last_name", "street", "city"]
dfs['customers'][cols] = (
    dfs['customers'][cols]
        .apply(lambda s: s.str.lower().str.replace(" ", "_", regex=False))
)

dfs['customers']['street'] = (
    dfs['customers']['street']
        .str.replace(r"[^\w\s]", "", regex=True)    
        .str.replace(r"_+$", "", regex=True)        
)
```

```python
col = ["first_name" ,"last_name"  ]
dfs['staffs'][col]  = dfs['staffs'][col].apply(lambda x : x.str.lower() )
```

# MISSING 

```python

dfs['customers']['phone'] = dfs['customers']['phone'].fillna("unknown" )

```

```python
cols_miss_stores =['email']
dfs['stores'][cols_miss_stores] = dfs['stores'][cols_miss_stores].fillna("unknown" )
dfs['stores']['zip_code'] = dfs['stores']['zip_code'].fillna(-1 )
```

```python
str_cols_miss =['email'	, 'last_name' , 'phone']
dfs['staffs'][str_cols_miss] = dfs['staffs'][str_cols_miss].fillna("unknown" )

numeric_cols_miss =['store_id'	, 'manager_id']
dfs['staffs'][numeric_cols_miss] = dfs['staffs'][numeric_cols_miss].fillna(-1)


dfs['orders'] = dfs['orders'].dropna()
```

##  Clean multi-valued phone numbers

```python
def clean_phone(phone):
    if pd.isna(phone) or phone == "unknown":
        return "unknown"
    if ',' in str(phone):
        return str(phone).split(',')[0].strip()
    return str(phone)

for df_name in ['customers', 'stores', 'staffs']:
    if 'phone' in dfs[df_name].columns:
        dfs[df_name]['phone'] = dfs[df_name]['phone'].apply(clean_phone)
```

```python
def save_cleaned_data(dfs):
    output_dir = "cleaned_data"
    os.makedirs(output_dir, exist_ok=True)
    
    for name, df in dfs.items():
        file_path = os.path.join(output_dir, f"{name}.csv")
        df.to_csv(file_path, index=False)
        print(f"Saved {name} to {file_path}")
```
