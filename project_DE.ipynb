## modify words

```python
def clean_text_columns(df):
    text_cols = df.select_dtypes(include="object").columns

    df[text_cols] = (
        df[text_cols].apply(lambda col:col.str.lower()
                                        .str.strip()
                                       .str.replace(" ", "_", regex=False)
            )
    )
    return df
```

```python
def clean_address(series):
    return (
        series.str.lower()
              .str.replace(" ", "_", regex=False)
              .str.replace(r"[^\w\s]", "", regex=True) # delete . 
              .str.replace(r"_+$", "", regex=True)  #delete final _
    )

```

```python
# it handle all phone and fill miss
def clean_phone(series):
    return (
        series.fillna("unknown") # handle missing
              .astype(str)
              .str.replace(r"[^\d]", "", regex=True)  #delete characters
              .replace("","unknown")  # if the num is char only and become empty to fill
              
    )
    
  
```

```python
def remove_negative(series):
    return(
        
        series[series>=0]
    )

```

```python

for name, df in dfs.items():
    df = clean_text_columns(df)

    if name == "customers" and "street"    in df.columns:
        df["street"] = clean_address(df["street"])
        
    if name == "products" and "product_name"   in df.columns:
        df["product_name"] = clean_address(df["product_name"]).str.replace(r"_+", "_", regex=True)
        
    if name == "staffs" and "phone" in df.columns:
        df["phone"] = clean_phone(df["phone"] )
        
    if name == "stores" and "phone" in df.columns:
        df["phone"] = clean_phone(df["phone"] )
        
    if name == "customers" and "phone" in df.columns:
        df["phone"] = clean_phone(df["phone"] )
        
    if name == "order_items" and "quantity" in df.columns:
        df["quantity"] = remove_negative(df["quantity"] )
    
    if name == "stocks" and "quantity" in df.columns:
        df["quantity"] = remove_negative(df["quantity"] )  
        
    dfs[name] = df

```

# MISSING 

```python
def show_missing():
  for name, df in dfs.items():
    null_count = df.isna().sum()                
    total_null = null_count.sum()               
    
    if total_null > 0:
        print(f"{name}: Total missing = {total_null}")
        print(null_count[null_count > 0])      
  if total_null == 0:
    print("NOT have missing value")
show_missing()
```

```python
 
# convert to datetime and handle missing data by default date 
date_cols = ['order_date' , 'required_date' , 'shipped_date']
for col in date_cols:
   dfs['orders'][col] = pd.to_datetime(dfs['orders'][col], errors='coerce')
   
dfs['orders'] = dfs['orders'].fillna(pd.Timestamp("2080-01-01"))

# fill stores
cols_miss_stores =['email']
dfs['stores'][cols_miss_stores] = dfs['stores'][cols_miss_stores].fillna("unknown" )

dfs['stores']['zip_code'] = dfs['stores']['zip_code'].fillna(-1 )


#staffs
str_cols_miss =['email'	, 'last_name' ]
dfs['staffs'][str_cols_miss] = dfs['staffs'][str_cols_miss].fillna("unknown" )




```

#  Data Transformation

## Merge products with brands and categories to reduce joins

```python

temp_df = dfs['products'].merge(dfs['brands'], on='brand_id', how='inner')
products = temp_df.merge(dfs['categories'], on='category_id', how='inner')

```

```python
tables = ['staffs' , 'customers']
for table in tables :
    df = dfs[table]
    df['full_name'] = df['first_name'] + '_' + df['last_name']
        
    dfs[table] = df
       
```

```python
df = dfs['order_items']
#Calculate total price per order item
df.loc[:, 'total_price'] = (df['list_price'] * df['quantity'] * (1 - df['discount'])).round(2) 

df_order_total_amount  =( df.groupby(by='order_id').agg(
    counts_items=('item_id', 'count') , 
    sums_money = (('total_price' , 'sum'))
    )
                          ).round(2).reset_index()




dfs['order_totals'] = df_order_total_amount


```

```python
import pandas as pd
import os
import json

# input folder (CSV files)
csv_folder = r"F:\DE_Final\analysis1"

# output json file
output_json = r"F:\DE_Final\frontend_data.json"

data_for_frontend = {}

for file_name in os.listdir(csv_folder):
    if file_name.endswith(".csv"):
        file_path = os.path.join(csv_folder, file_name)
        table_name = os.path.splitext(file_name)[0]

        df = pd.read_csv(file_path , nrows = 50)

        # convert dataframe to list of dicts
        data_for_frontend[table_name] = df.to_dict(orient="records")

        print(f"Loaded: {table_name}")

# write one json file
with open(output_json, "w", encoding="utf-8") as f:
    json.dump(data_for_frontend, f, indent=4, ensure_ascii=False)

```
